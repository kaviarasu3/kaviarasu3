```python
def register():
    db = open("/content/email_register.txt", "r")
    username = input("Create your username in format __@__.__: ")
    a = []
    for line in db:
        x = line.split(",")
        a.append(x[0])

    if username in a:
        print("Try again with different username")
        register()

    elif username.count('@') != 1 and username.count('.') != 1:
        print("enter username in proper format")
        register()

    elif ((username.index('@')) - (username.index('.'))) == 1:
        print("Enter properly please!")
        register()

    elif username[0] in range(0, 9):
        print("Start with alphabet characters only")
        register()

    elif (username[0] == '@' or username[0] == '$' or username[0] == '_' or username[0] == '%' or username[0] == '!' or username[0] == '#' or username[0] == '*' or username[
        0] == '&'):
        print("Start with only characters")
        register()

    else:
        print("Username created")

    password = input("Create your password with atleast one capital letter one integer and one special character: ")
    s = False

    if len(password) < 5 and len(password) > 16:
        print("Create Password with length between 5 an 16, Try again")
        register()

    if len(password) > 5 and len(password) < 16:
        digit = 0 
        lower = 0 
        upper = 0 
        specialchar = 0
        for i in password:
            if i.isdigit():
                digit += 1
            if i.islower():
                lower += 1
            if i.isupper():
                upper += 1
            if (i == '@' or i == '$' or i == '_' or i == '%' or i == '!' or i == '#' or i == '*' or i == '&'):
                specialchar += 1
            if (lower >= 1 and upper >= 1 and specialchar >= 1 and digit >= 1 and lower + specialchar + upper + digit == len(password)):
                s = True

    if s:
        b = input("Confirm Password: ")
        while (b != password):
            print("Password not match, Try again")
            b = input("Try again: ")

    else:
        print("Try again")
        register()

    file = open("/content/email_register.txt", "a")
    file.write(username + "," + password + "\n")
    file.close()
    login()

def login():
    X=input("Enter your username to login: ")
    X = X.strip()
    db = open("/content/email_register.txt", "r")
    c = []
    for line in db:
        x = line.split(",")
        c.append(x[0])

    if X in c:
        Y=input("Please Enter your password: ")
        Y=Y.strip()
        file1=open("/content/email_register.txt","r").readlines()
        for x in file1:
            x=x.strip()
            info=x.split(",")
            if X==info[0] and Y==info[1]:
                print("Welcome to mail homepage")
                break
            else:
                Forgot_password = input("Forgot Password [Yes/No] : ")

                if Forgot_password == "No":
                    print("try")
                    login()

                if Forgot_password == "Yes":
                    password = input("Create new password with atleast one capital letter one integer and one special character: ")
                    s = False

                    if len(password) < 5 and len(password) > 16:
                        print("Create Password with length between 5 an 16, Try Again")
                        register()

                    if len(password) > 5 and len(password) < 16:
                        lower = 0 
                        upper = 0 
                        specialchar = 0 
                        digit = 0
                        for i in password:
                            if i.isdigit():
                                digit += 1
                            if i.islower():
                                lower += 1
                            if i.isupper():
                                upper += 1
                            if (i == '@' or i == '$' or i == '_' or i == '%' or i == '!' or i == '#' or i == '*' or i == '&'):
                                specialchar += 1
                            if (lower >= 1 and upper >= 1 and specialchar >= 1 and digit >= 1 and lower + specialchar + upper + digit == len(password)):
                                s = True

                    if s:
                        b = input("Confirm Password: ")
                        while (b != password):
                            print("Password not match, Try Again")
                            b = input("Try Again: ")

                    else:
                        print("Sorry,Try again to login")
                        login()

                    file = open("/content/email_register.txt", "w")
                    file.write(X + "," + password + "\n")
                    file.close()

    else:
        print("Unregister user you need to register first")
        register()

def welcome():
    print("Homepage")
    print("For login click Login and if new user click register")
    W=input("Login|Register[L/R]: ")
    if W=="L":
        login()
    elif W=="R":
        register()
    else:
        print("Enter in proper manner only please")
        welcome()
welcome()
```
